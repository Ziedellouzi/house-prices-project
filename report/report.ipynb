## Transformation of the Target Variable

The target variable `SalePrice` exhibits a strong right-skewed distribution, as shown by its histogram and skewness coefficient. This indicates the presence of a long tail with a small number of houses having very high prices, which can negatively affect statistical modeling.

Many statistical methods used in this project, such as ANOVA, linear regression, and hypothesis testing, rely on assumptions of normality and constant variance. A highly skewed distribution violates these assumptions and may lead to unreliable estimates and biased results.

To address this issue, a logarithmic transformation was applied to the target variable using the `log1p` function:

$$
\text{LogSalePrice} = \log(\text{SalePrice} + 1)
$$

This transformation has several advantages:

- It reduces the skewness of the distribution, making it more symmetric and closer to a normal distribution.
- It compresses the scale of large values, thereby reducing the influence of outliers.
- It stabilizes the variance, improving the performance and reliability of statistical models.

After transformation, the skewness of the distribution significantly decreased, indicating a more suitable shape for further analysis.

In addition, this transformation is consistent with the evaluation metric used in the Kaggle competition, which computes the Root Mean Squared Error (RMSE) on the logarithm of predicted and actual prices.

Therefore, all subsequent analyses in this project are conducted using the transformed variable `LogSalePrice`.
